```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from keras.models import Sequential
from keras.layers import Dense ,Activation
```

```python
# upload and read the data
from google.colab import files
uploaded = files.upload()
```

```python
# load the dataset
import io
df = pd.read_csv(io.BytesIO(uploaded['Salary.csv']))
```

```python
df.info()
```

```python
# slicing the data
x = df.iloc[:,0]
y = df.iloc[:,1]
x
```

```python
y
```

```python
sns.lmplot(x='YearsExperience',y='Salary',data=df)
```

```python
# Start to build he model
model = Sequential()
model.add(Dense(1,input_dim=1))
model.add(Activation('linear'))

model.summary()
```

```python
model.compile(loss='mse',optimizer='sgd', metrics=['mse'])
```

```python
# training the model
history = model.fit(x,y,epochs=600,verbose=1)
```

```python
#start to use the model
pred = model.predict(x)
pred
```

```python
# Visualization  data
plt.scatter(x,y,c="blue")
#plt.plot(x,pred,coloe="g")
plt.plot(x, pred, color='green')
plt.show()
```
